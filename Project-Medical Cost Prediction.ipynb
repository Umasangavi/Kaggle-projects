# Medical Cost  Prediction

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet,RidgeCV,LassoCV,ElasticNetCV
```

```python
df=pd.read_csv('insurance.csv')
df
```

```python
df.describe(include='all')
```

```python
df.info()
```

```python
print("Number of null values:",df.isnull().sum().sum())
```

```python
print(df.nunique())
```

```python
df["sex"]=df["sex"].replace({'male':1,'female':0})
print(df["sex"])
```

```python
df["smoker"]=df["smoker"].replace({'No':0,'Yes':1})
print(df)
```

```python
df["region"].value_counts()
```

```python
df["region"]=df["region"].replace({'southeast':0,'southwest':1,'northwest':2,'northeast':3})
df
```

```python
y=df["charges"].copy()
X=df.drop("charges" ,axis=1).copy()
X 
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,train_size=0.7,shuffle=True,random_state=4)
X_train
```

```python
y_test
```

```python
X_train,X_test,y_train,y_test
```

```python
model_1=LinearRegression()
model_1.fit(X_train,y_train)
model_1.score(X_test,y_test)*100
```

```python
model_2=Ridge()
model_2.fit(X_train,y_train)
model_2.score(X_test,y_test)*100
```

```python
model_3=Lasso()
model_3.fit(X_train,y_train)
model_3.score(X_test,y_test)*100
```

```python
model_4=ElasticNet()
model_4.fit(X_train,y_train)
model_4.score(X_test,y_test)*100
```

```python
model_5=RidgeCV()
model_5.fit(X_train,y_train)
model_5.score(X_test,y_test)*100
```

```python
model_6=LassoCV()
model_6.fit(X_train,y_train)
model_6.score(X_test,y_test)*100
```

```python
model_7=ElasticNetCV()
model_7.fit(X_train,y_train)
model_7.score(X_test,y_test)*100
```

```python
print(model_1.score(X_test,y_test)*100)
print(model_2.score(X_test,y_test)*100)
print(model_3.score(X_test,y_test)*100)
print(model_4.score(X_test,y_test)*100)
print(model_5.score(X_test,y_test)*100)
print(model_6.score(X_test,y_test)*100)
print(model_7.score(X_test,y_test)*100)
```
